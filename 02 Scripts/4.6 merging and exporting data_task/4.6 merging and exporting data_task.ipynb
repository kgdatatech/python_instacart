#### 2. Import the orders_products_combined dataframe from the pickle file

```python
# Import the orders_products_combined dataframe from the pickle file

df_ords_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
# Import the products_checked dataframe from the csv file

df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
```

#### 4. Determine a suitable way to combine the orders_products_combined dataframe with the products data set

```python
# Combine the orders_products_combined dataframe with the products data set using inner join

df_ords_prods_merge = df_ords_combined.merge(df_prods, on = 'product_id')
```

```python
# Check the output merged dataframe 

df_ords_prods_merge.head()
```

#### 5. Confirm the results of the merge using the merge flag

```python
# Confirm results with merge flag

df_ords_prods_merge = df_ords_combined.merge(df_prods, on = 'product_id', indicator = True)
```

```python
# Check the output merged dataframe 

df_ords_prods_merge.head()
```

```python
# Sum up all the values in the “_merge” column

df_ords_prods_merge['_merge'].value_counts()
```

```python
# Check # Combine the orders_products_combined dataframe with the products data set using outer join

df_ords_prods_merge1 = df_ords_combined.merge(df_prods, on = 'product_id', how = 'outer')
```

```python
df_ords_prods_merge1
```

```python
# Remove redundant index column 'unnamed: 0'

df_ords_prods_merge = df_ords_prods_merge.drop(columns=['Unnamed: 0'])
```

```python
# Check output for successful removal 

df_ords_prods_merge.head()
```

#### 6. Export the newly created dataframe as orders_products_merged in a suitable format (taking into consideration the size)

```python
# Export data to pkl

df_ords_prods_merge.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.pkl'))
```

 #### If-Statements with User-Defined Functions

```python
# Define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#  Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check the values in new column

df['price_range'].value_counts(dropna = False)
```

```python
# Check max value of subset

df['prices'].max()
```

#### 4. If-Statements with the loc() Function

```python
# Create conditions (1)

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Create conditions (2)

df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
# Create conditions (3)

df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Create conditions for entire dataframe (1)

ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Create conditions for entire dataframe (2)

ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
# Create conditions for entire dataframe (3)

ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
ords_prods_merge['price_range_loc'].value_counts(dropna = False)
```

#### 5. If-Statements with For-Loops

```python
# Check orders frequency 

ords_prods_merge['order_day_of_week'].value_counts(dropna = False)
```

```python
# Create a new column in your ords_prods_merge dataframe that summarizes how busy each day of the week is

result = []

for value in ords_prods_merge["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create a new column within your ords_prods_merge dataframe and set it equal to result

ords_prods_merge['busiest_day'] = result
```

```python
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

#### 6. Exercise 4.7 task questions

```python
# Q2
# Create a new column in your ords_prods_merge dataframe that summarizes the 2 busiest and slowest days of the week

result_new = []

for value in ords_prods_merge["order_day_of_week"]:
  if value in [0, 1]:
    result_new.append("Busiest days")
  elif value in [3, 4]:
    result_new.append("Slowest days")
  else:
    result_new.append("Regularly busy")
```

##### note:
used a list to check for multiple days

```python
result_new
```

```python
# Create a new column within your ords_prods_merge dataframe and set it equal to result_new

ords_prods_merge['busiest_days'] = result_new
```

```python
# Q3
# Print the frequency of result_new

ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
# Q4
# Identify the busiest hours of the day

# Check order_hour frequency 

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

##### note:
high peak = 10am,
most orders = 8am-7pm,
average orders = 8pm-1am, 6am-7am,
fewest orders = 3am-5am, 
low peak = 3am

```python
# Create a new column containing these labels called “busiest_period_of_day.”
# label as “Most orders,” “Average orders,” and “Fewest orders.” 


result_hour = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9, 17, 8, 18, 19]:
    result_hour.append("Most orders")
  elif value in [5, 2, 4, 3]:
    result_hour.append("Fewest orders")
  else:
    result_hour.append("Average orders")
```

```python
result_hour
```

```python
# Create a new column within your ords_prods_merge dataframe and set it equal to result_hour

ords_prods_merge['busiest_period_of_day'] = result_hour
```

```python
# Q5
# Print the frequency of result_hour

ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

#### Q5. Fundamental data quality and consistency checks

```python
# Data consistency checks
df_cust.describe()
```

```python
# Check for mixed types
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].map(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

```python
df_cust.info()
```

```python
# Change data type of first_name column to 'str'
df_cust['first_name'] = df_cust['first_name'].astype('str')
```

```python
# Find missing values
df_cust.isnull().sum()
```

```python
# Finding duplicates
df_cust[df_cust.duplicated()]
```

```python
# Check for mixed types (recheck)
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].map(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

#### Q6. Combine customer data with the rest of the prepared Instacart data

```python
# Check df_cust dataframe
df_cust.head()
```

```python
# Check df_prep dataframe
df_prep.head()
```

```python
df_prep.dtypes
```

```python
# Change data types for df_prep order_id and user_id columns to match for successful merge
df_prep['user_id'] = df_prep['user_id'].astype('str')
```

```python
df_prep['user_id'].dtype
```

```python
# Combine the df_prep dataframe with the df_cust data set using inner join
df_merge = df_prep.merge(df_cust, on = 'user_id')
```

```python
# Check the output merged dataframe 
df_merge.head()
```

```python
df_merge.tail()
```

```python
# Confirm results with merge flag
df_merge = df_prep.merge(df_cust, on = 'user_id', indicator = True)
```

```python
# Check the output merged dataframe 
df_merge.head()
```

```python
# Sum up all the values in the “_merge” column
df_merge['_merge'].value_counts()
```

#### note:
"both" indicates a successful merge of both dataframes.
